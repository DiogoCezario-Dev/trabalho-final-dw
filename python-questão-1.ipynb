```python
# terceira questão #

class Documento:
    def render(self):
        raise NotImplementedError("Subclasses devem implementar render().")

class Relatório(Documento):
    def __init__(self, titulo, conteudo):
        self.titulo = titulo
        self.conteudo = conteudo
    
    def render(self):
        return "[RELATÓRIO] Título: {self.titulo} Conteúdo: {self.conteudo}"    

class NotaFiscal(Documento):
    def __init__ (self, numero, valor):
        self.numero = numero 
        self.valor = valor
    
    def render(self):
        return "[NOTA FISCAL]\nNúmero: {self.numero}\nValor: R${self.valor:.2f}"


    class Carta(Documento):
        def __init__ (self, destino, mensagem):
            self.destino = destino
            self.mensagem = input("digite na carta")

        def render(self):
            return "[CARTA]\nPara: {self.destinatario}\nMensagem: {self.mensagem}"
    
    def processar_documentos(lista_de_documentos):
        resultados = []
        quantidade = {}

        for doc in lista_de_documentos: 
            resultados.append(doc.render())

            nome_tipo = doc.__class__.__name__
            quantidade[nome_tipo] = quantidade.get(nome_tipo, 0) + 1

        return resultados, quantidade
```
